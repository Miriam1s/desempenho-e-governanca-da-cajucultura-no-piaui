1. INTRODUÇÃO AO CAPÍTULO

Este capítulo consolida os achados empíricos obtidos ao longo da investigação, integrando as análises desenvolvidas nos Cadernos 01 a 04 em uma interpretação sistêmica do desempenho da cajucultura no estado do Piauí. O objetivo central é sintetizar evidências quantitativas e qualitativas, articulando-as com a literatura especializada e com uma perspectiva administrativa orientada à governança produtiva e ao desenvolvimento regional.

A análise permite avançar da descrição estatística para a compreensão estrutural da atividade, examinando sua dinâmica histórica, organização territorial, eficiência econômica e implicações institucionais.

2. SÍNTESE DOS RESULTADOS EMPÍRICOS
2.1 Dinâmica Histórica e Volatilidade Produtiva

A série histórica estadual (1974–2024) revelou forte volatilidade interanual da produção de castanha de caju no Piauí, com variações superiores a 200% entre anos de pico e anos de retração. Observam-se períodos em que a produção supera 50 mil toneladas, intercalados com anos em que o volume recua para patamares inferiores a 15 mil toneladas.

Essa instabilidade evidencia elevada dependência de fatores climáticos e limitada capacidade de mitigação de riscos estruturais, característica recorrente em sistemas agrícolas do semiárido nordestino (BUAINAIN et al., 2014). Do ponto de vista administrativo, tal volatilidade compromete previsibilidade de receita, planejamento de investimentos e estabilidade da renda agrícola.

2.2 Concentração Territorial da Produção

A análise da distribuição municipal demonstrou padrão altamente concentrado. Aproximadamente 20% dos municípios concentram cerca de 80% da produção estadual, configurando estrutura próxima ao princípio de Pareto.

Além disso:

Um único município apresenta produção acumulada superior a 6 milhões de unidades, destacando-se significativamente dos demais;

A maioria dos municípios apresenta produção inferior a 100 mil unidades;

Mais de 200 municípios compartilham menos de 20% do volume total produzido.

Essa configuração evidencia estrutura territorial dual: polos produtivos consolidados coexistem com ampla periferia produtiva de baixa intensidade.

Segundo Porter (1998), a formação de clusters pode gerar ganhos de eficiência coletiva; contudo, Fujita, Krugman e Venables (1999) alertam que estruturas excessivamente concentradas ampliam vulnerabilidades sistêmicas e dependência territorial.

2.3 Escala Produtiva e Captura de Valor

A tipologia econômica municipal revelou:

- 36,0% dos municípios combinam alto volume e alto valor;
- 35,6% apresentam baixo volume e baixo valor;
- 14,2% produzem muito, mas capturam pouco valor;
- 14,2% produzem pouco, mas capturam alto valor relativo.

A presença de 28,4% de municípios com desalinhamento estrutural entre produção e valor indica que a escala produtiva não é determinante isolada do desempenho econômico.

O ranking cruzado entre produção total e eficiência econômica revelou discrepâncias superiores a 60 posições, chegando a aproximadamente 70 posições em casos extremos. Municípios líderes em volume produtivo não figuram, necessariamente, entre os mais eficientes economicamente.

Esse resultado confirma que a captura de valor depende de fatores organizacionais e institucionais, conforme argumentado na literatura sobre governança das cadeias globais de valor (GEREFFI; HUMPHREY; STURGEON, 2005).

2.4 Produtividade Física e Eficiência Econômica

A análise da relação entre produtividade (kg/ha) e valor por hectare evidenciou ausência de correlação linear forte.

Constata-se que:

- Municípios com produtividade inferior a 1 kg/ha alcançam valores superiores a R$ 0,45 por hectare;
- Municípios com produtividade superior a 10 kg/ha registram retornos inferiores a R$ 0,20 por hectare;
- Casos extremos apresentam produtividade acima de 30 kg/ha com valor econômico próximo a zero.

Esse padrão reforça que eficiência econômica está mais associada à organização produtiva, inserção mercadológica e estrutura de custos do que à produtividade física isolada.

3. COMPARAÇÃO INTERESTADUAL: CONTEXTO REGIONAL

Segundo a PAM/IBGE (2023–2024):

- A Região Nordeste responde por mais de 95% da produção nacional de castanha de caju;
- O Ceará concentra aproximadamente 45–50% da produção brasileira;
- O Rio Grande do Norte responde por cerca de 15–20%;
- O Piauí participa com aproximadamente 10–15%.

A diferença central, entretanto, reside na estrutura da cadeia produtiva:

| Estado | Volume | Beneficiamento | Inserção Exportadora |
| ------ | ------ | -------------- | -------------------- |
| Ceará  | Alta   | Consolidado    | Elevada              |
| RN     | Média  | Intermediário  | Moderada             |
| Piauí  | Média  | Limitado       | Reduzida             |

A Embrapa (2020) destaca que a agregação de valor via beneficiamento pode elevar em até 60% o valor final da produção. A menor presença industrial no Piauí ajuda a explicar parte do desalinhamento entre escala e eficiência observado nos dados.

4. DISCUSSÃO: GOVERNANÇA, GESTÃO E POLÍTICAS PÚBLICAS

Os resultados indicam que a cajucultura piauiense enfrenta desafios estruturais que transcendem a dimensão produtiva.

A ausência de convergência econômica entre municípios sugere que políticas públicas podem não estar promovendo difusão homogênea de ganhos tecnológicos e institucionais. Estudos do IPEA (2021) apontam concentração do crédito rural em municípios previamente estruturados, o que tende a reforçar desigualdades regionais.

Além disso, a literatura de Schultz (1964) enfatiza que a transformação agrícola depende da eficiência alocativa e da adaptação institucional. Os dados indicam que municípios com menor escala, mas maior eficiência, apresentam melhor organização produtiva ou inserção mercadológica.

Do ponto de vista administrativo, os achados evidenciam fragilidade na governança da cadeia produtiva estadual, com limitações na coordenação vertical, agregação de valor e integração logística.

5. IMPLICAÇÕES ESTRATÉGICAS

Com base na análise integrada, identificam-se cinco eixos estratégicos prioritários:

5.1 Fortalecimento da Governança da Cadeia

Integração entre produtores, processadores e mercados para ampliar captura de valor.

5.2 Industrialização Regional

Implantação de polos de beneficiamento para reduzir evasão de renda.

5.3 Política de Eficiência Econômica

Substituir foco exclusivo em volume por metas de valor por hectare.

5.4 Difusão Tecnológica Territorializada

Ampliar adoção de clones e manejo tecnificado.

5.5 Estratégia Exportadora Estadual

Reduzir dependência de intermediários externos.

6. CONCLUSÃO GERAL

A cajucultura piauiense caracteriza-se por:

- Alta volatilidade histórica;
- Forte concentração territorial;
- Heterogeneidade econômica persistente;
- Desalinhamento entre escala produtiva e eficiência;
- Fragilidade na governança institucional.

Os resultados demonstram que produzir mais não é suficiente para gerar maior renda. O fortalecimento sustentável da atividade requer abordagem sistêmica, integrando gestão, coordenação institucional e políticas públicas orientadas à eficiência econômica e agregação de valor.

A pesquisa confirma empiricamente que os determinantes do desempenho da cajucultura no Piauí são predominantemente organizacionais e institucionais, e não apenas produtivos.
