1. Contextualização e Delimitação do Estudo

Este estudo está vinculado ao Projeto de Pesquisa PV18851-2025, intitulado “Análise do Crescimento da Cajucultura no PIB Brasileiro: uma Perspectiva Administrativa”, em execução no âmbito da Universidade Estadual do Piauí (UESPI).

A cajucultura, centrada no cultivo do cajueiro (Anacardium occidentale L.), possui elevada relevância socioeconômica no Brasil, especialmente na região Nordeste, com destaque para os estados do Ceará, Piauí e Rio Grande do Norte. Além de sua importância produtiva, a atividade exerce papel estratégico na geração de renda, emprego e dinamização das economias rurais do semiárido.

Apesar da expressiva área plantada e da relevância territorial da cultura no estado do Piauí, observa-se que o desempenho produtivo e econômico ainda se encontra aquém de seu potencial, sobretudo em função de limitações administrativas, baixa adoção de práticas gerenciais, fragilidades logísticas, restrições ao crédito e insuficiência de políticas públicas eficazes.

Nesse contexto, o presente estudo adota uma perspectiva administrativa, buscando compreender como fatores de gestão, organização produtiva e ambiente institucional influenciam o desempenho da cajucultura e sua contribuição para o Produto Interno Bruto (PIB) agroindustrial.

2. Problema de Pesquisa

Como os fatores administrativos, produtivos e institucionais influenciam o desempenho econômico da cajucultura no Brasil, especialmente no estado do Piauí, e de que forma esses elementos impactam sua contribuição para o PIB agroindustrial?

3. Objetivo Geral

Analisar o desempenho da cajucultura no contexto econômico do Brasil, com foco no estado do Piauí, considerando sua contribuição produtiva e econômica e os fatores administrativos que impactam sua cadeia produtiva.

4. Objetivos Específicos

- Avaliar a evolução da produção e da comercialização do caju e da castanha de caju no Brasil ao longo dos últimos anos;

- Identificar os estados com maior participação na cajucultura e seus respectivos resultados econômicos;

- Investigar de que forma fatores administrativos, como gestão da produção, logística e estratégias de comercialização, afetam o desempenho do setor;

- Analisar o papel das políticas públicas, incentivos e investimentos no fortalecimento, estabilidade e competitividade da cajucultura, com ênfase no estado do Piauí.

5. Caracterização Metodológica
5.1 Natureza e Tipo de Pesquisa

A pesquisa caracteriza-se como aplicada, de natureza descritiva e analítica, alinhada à área de conhecimento da Administração, com foco em setores produtivos específicos.

5.2 Abordagem Metodológica

Adota-se uma abordagem mista, combinando:

Pesquisa quantitativa, baseada na análise de dados secundários oficiais referentes à produção, área plantada, produtividade, valor da produção, eficiência econômica e estabilidade produtiva;

Pesquisa qualitativa, voltada à interpretação dos resultados à luz da literatura acadêmica, documentos institucionais e estudos sobre gestão, políticas públicas e organização da cadeia produtiva da cajucultura.

5.3 Fonte e Base de Dados

Os dados quantitativos utilizados são provenientes, majoritariamente, da Pesquisa Agrícola Municipal (PAM/IBGE), complementados por informações técnicas e institucionais de órgãos como EMBRAPA e FAO, conforme descrito no projeto de pesquisa.

No total, são analisados 11 conjuntos de dados, abrangendo escalas nacional, estadual e municipal, com especial atenção ao estado do Piauí.
