## 1. Contextualização do Estudo

A cajucultura desempenha papel relevante na economia agrícola brasileira, especialmente na região Nordeste, onde concentra a maior parte da produção nacional de castanha de caju. Além de sua importância econômica, a atividade possui expressiva dimensão social, contribuindo para a geração de emprego e renda em áreas caracterizadas por limitações climáticas e estruturais.

Apesar dessa relevância, observa-se que o desempenho econômico da cajucultura nem sempre reflete seu potencial produtivo, sobretudo em estados como o Piauí. Fatores relacionados à gestão, organização produtiva, acesso a políticas públicas e condições climáticas influenciam diretamente os resultados do setor.

Nesse contexto, a presente pesquisa adota uma **perspectiva administrativa**, buscando compreender como aspectos econômicos, produtivos e climáticos se relacionam com a evolução da cajucultura e sua contribuição para o Produto Interno Bruto (PIB) agropecuário.

## 2. Objetivo Geral

Analisar o desempenho da cajucultura no contexto econômico brasileiro, com ênfase no estado do Piauí, considerando sua contribuição para o PIB agropecuário e os fatores administrativos e climáticos que influenciam sua cadeia produtiva.

## 4. Delineamento da Pesquisa e Abordagem Metodológica

O presente estudo caracteriza-se como uma **pesquisa aplicada**, de natureza **quantitativa**, com objetivos **descritivos e exploratórios**. A escolha desse delineamento justifica-se pela necessidade de compreender o comportamento econômico e produtivo da cajucultura, bem como identificar padrões, tendências e associações entre variáveis produtivas, econômicas e climáticas.

Do ponto de vista do método científico, adota-se uma **abordagem empírico-analítica**, baseada na utilização de **dados secundários oficiais**, amplamente empregados em estudos econômicos e administrativos do setor agropecuário.

A análise é conduzida em diferentes escalas territoriais — nacional, estadual e municipal — permitindo uma visão hierarquizada do fenômeno estudado, conforme recomendado em pesquisas aplicadas na área de Administração e Economia Agrícola.

## 5. Fontes de Dados e Construção da Base Empírica

A base empírica do estudo foi construída a partir da integração de dados produtivos, econômicos e climáticos, provenientes de fontes oficiais, garantindo confiabilidade, comparabilidade e reprodutibilidade dos resultados.

### 5.1 Pesquisa Agrícola Municipal (PAM/IBGE)

Os dados referentes à produção agrícola foram extraídos da **Pesquisa Agrícola Municipal (PAM)**, realizada pelo Instituto Brasileiro de Geografia e Estatística (IBGE). A PAM constitui a principal fonte oficial de informações sobre a produção agrícola brasileira em nível municipal, sendo amplamente utilizada em estudos acadêmicos e institucionais.

Foram considerados os seguintes indicadores:
- área colhida (hectares);
- quantidade produzida;
- rendimento médio da produção;
- valor da produção.

Em razão de alterações metodológicas ao longo da série histórica, especialmente no que se refere à nomenclatura dos produtos, as análises temporais priorizam o produto **castanha de caju**, assegurando homogeneidade conceitual e consistência estatística.



### 5.2 Dados Climáticos

As informações climáticas foram obtidas a partir de bases meteorológicas oficiais, organizadas por estado da região Nordeste, no período de 2000 a 2025. Os dados contemplam variáveis fundamentais para a análise agroclimática, tais como:

- precipitação total;
- temperatura média do ar;
- umidade relativa do ar.

Originalmente disponibilizados em escala horária, os dados climáticos passaram por processo de agregação para a escala anual, procedimento metodologicamente adequado para compatibilização com os dados produtivos da PAM.


### 5.2 Dados Climáticos

As informações climáticas foram obtidas a partir de bases meteorológicas oficiais, organizadas por estado da região Nordeste, contemplando o período de 2000 a 2025. As variáveis consideradas incluem, principalmente:

- precipitação total;
- temperatura média do ar;
- umidade relativa do ar.

Os dados climáticos originalmente registrados em escala horária foram agregados para a escala anual, possibilitando sua integração com os dados produtivos da PAM.


## 6. Procedimentos Analíticos

A estratégia analítica adotada neste estudo estrutura-se em etapas complementares, permitindo uma compreensão progressiva do fenômeno investigado.

### 6.1 Análise Descritiva

A análise descritiva tem por finalidade caracterizar a evolução da cajucultura, por meio de estatísticas básicas, tais como médias, totais e taxas de variação. Nessa etapa, analisam-se a produção, o valor econômico e os indicadores de produtividade, tanto em nível nacional quanto estadual e municipal.

### 6.2 Análise Exploratória

A análise exploratória busca identificar padrões espaciais e estruturais da produção, com ênfase na:
- especialização produtiva municipal;
- concentração espacial da produção;
- eficiência econômica, mensurada a partir da relação entre valor gerado e área colhida.

Essas análises permitem avaliar a heterogeneidade do desempenho da cajucultura e discutir implicações administrativas e gerenciais.

### 6.3 Análise de Correlação Climática

Com o objetivo de investigar a influência das condições climáticas sobre a produção de castanha de caju, emprega-se a análise de correlação entre variáveis climáticas (especialmente precipitação) e indicadores produtivos.

Utilizam-se coeficientes de correlação apropriados, como Pearson e Spearman, de acordo com a distribuição dos dados. Ressalta-se que essa etapa possui caráter exploratório, não implicando inferência causal.

## 7. Limitações e Considerações Metodológicas

Algumas limitações metodológicas devem ser reconhecidas. Primeiramente, o uso de dados secundários implica dependência da qualidade e da metodologia adotada pelas instituições responsáveis pela coleta. Ademais, a agregação dos dados climáticos em escala estadual pode atenuar variações microclimáticas relevantes.

Ainda assim, tais limitações não comprometem os objetivos do estudo, uma vez que a pesquisa busca identificar tendências gerais e relações estruturais, e não realizar previsões pontuais. A adoção de análises complementares, como a incorporação de variáveis climáticas, amplia a robustez interpretativa dos resultados e contribui para uma compreensão mais abrangente da dinâmica da cajucultura.
