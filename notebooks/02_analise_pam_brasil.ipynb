Este notebook tem como objetivo analisar o **desempenho da cajucultura no Brasil**, considerando a evolução da produção e do valor econômico, bem como a participação dos estados produtores. As análises aqui desenvolvidas estão diretamente associadas aos **objetivos específicos 1 e 2** da pesquisa.

## 1. Objetivo da Análise

- Avaliar a evolução da produção e do valor econômico da cajucultura no Brasil;
- Identificar os estados com maior participação na produção de caju e castanha de caju.



```python
# 2. Importação das bibliotecas
# Esta etapa garante reprodutibilidade e padronização das análises

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 6)

```

```python
# 3. Carregamento dos dados
serie_brasil = pd.read_csv(
    "../data/raw/Pesquisa Agrícola Municipal (PAM)/1. serie_historica_caju_castanha_brasil.csv"
)
serie_brasil.head()
```

```python
# Produção total da cajucultura (caju + castanha)
cajucultura_brasil = (
    serie_brasil
    .groupby("ano", as_index=False)["producao_total"]
    .sum()
)
```

```python
cajucultura_brasil.head(10)

```

```python
import plotly.express as px

fig = px.line(
    cajucultura_brasil,
    x="ano",
    y="producao_total",
    title="Evolução da Produção da Cajucultura no Brasil",
    labels={
        "ano": "Ano",
        "producao_total": "Produção total da cajucultura"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()

```

Interpretação da evolução da produção da cajucultura no Brasil

A visualização interativa da evolução da produção da cajucultura no Brasil evidencia uma trajetória marcada por duas fases estatisticamente e metodologicamente distintas, cuja compreensão é fundamental para a correta interpretação dos dados.

Na primeira fase, que se estende até o final da década de 1980, observa-se um volume elevado de produção, com valores que atingem patamares superiores a 10 milhões de toneladas. Esse comportamento reflete o modelo de mensuração então adotado pela Pesquisa Agrícola Municipal (PAM), no qual o produto “caju” era registrado de forma agregada, incorporando o pseudofruto como principal referência produtiva. As oscilações interanuais expressivas observadas nesse período são compatíveis com a elevada sensibilidade da produção agrícola a fatores climáticos, tecnológicos e metodológicos, além de possíveis limitações nos processos de coleta e padronização estatística da época.

A partir do final dos anos 1980, verifica-se uma redução abrupta nos valores absolutos da produção, seguida por uma trajetória de menor magnitude e maior estabilidade relativa. Essa inflexão não deve ser interpretada como um colapso produtivo da cajucultura, mas como consequência direta de uma mudança conceitual e estatística promovida pelo IBGE, que passou a priorizar a castanha de caju como principal unidade de registro, em consonância com sua relevância econômica e comercial na cadeia produtiva.

Sob a perspectiva da análise integrada da cajucultura — entendida como uma única atividade produtiva — o gráfico revela que, embora a escala dos registros tenha sido substancialmente alterada, a produção mantém-se contínua ao longo do tempo, com variações compatíveis com o comportamento esperado de uma cultura fortemente dependente de condições climáticas e estruturais. A estabilidade relativa observada nas décadas mais recentes sugere um setor que, embora não apresente crescimento expressivo em termos físicos, mantém-se funcional e economicamente relevante, especialmente quando analisado em conjunto com indicadores de valor da produção.

Do ponto de vista metodológico, a adoção de uma série integrada da cajucultura permite contornar descontinuidades estatísticas e assegurar uma leitura mais coerente da trajetória histórica da atividade. Tal abordagem é particularmente relevante em estudos de natureza administrativa e econômica, nos quais o foco analítico recai sobre o desempenho da cadeia produtiva como um todo, e não sobre produtos isolados que representam apenas diferentes estágios ou formas de mensuração da mesma atividade agrícola.

Por fim, a utilização de uma visualização interativa amplia o potencial exploratório da análise, possibilitando a identificação de anos específicos com variações atípicas e favorecendo a transparência na comunicação dos resultados. Esse recurso reforça o papel da pesquisa acadêmica na tradução de dados complexos em informações acessíveis, contribuindo para a aproximação entre conhecimento científico, gestão pública e sociedade.

Observação metodológica complementar

A interpretação dos níveis absolutos de produção deve sempre ser acompanhada da compreensão das transformações metodológicas da base de dados. Assim, análises comparativas mais aprofundadas — especialmente aquelas que envolvem correlação com variáveis climáticas ou avaliação de eficiência econômica — devem privilegiar períodos metodologicamente homogêneos, assegurando maior robustez estatística aos resultados.

```python
# Valor total da cajucultura (caju + castanha)
valor_cajucultura_brasil = (
    serie_brasil
    .groupby("ano", as_index=False)["valor_total_producao"]
    .sum()
)
valor_cajucultura_brasil.head()
```

```python
import plotly.express as px

fig = px.line(
    valor_cajucultura_brasil,
    x="ano",
    y="valor_total_producao",
    title="Evolução do Valor da Produção da Cajucultura no Brasil",
    labels={
        "ano": "Ano",
        "valor_total_producao": "Valor total da produção (R$)"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()
```

# Interpretação da evolução do valor da produção da cajucultura no Brasil

- A evolução do valor da produção da cajucultura no Brasil revela uma trajetória distinta daquela observada para os volumes físicos produzidos, evidenciando a importância de se analisar a atividade sob a ótica econômica. Observa-se que, a partir da década de 1990, o valor da produção passa a ser registrado de forma mais consistente, acompanhando a consolidação da castanha de caju como principal produto econômico da cadeia produtiva.

- Ao longo do período analisado, verifica-se uma tendência geral de crescimento do valor da produção, ainda que marcada por oscilações interanuais. Essas variações refletem a sensibilidade da atividade a fatores conjunturais, como flutuações de preços, condições de mercado e possíveis choques externos, sem que isso implique necessariamente mudanças proporcionais na quantidade produzida.

- A partir dos anos 2000, o gráfico evidencia um processo de valorização gradual da cajucultura, com incrementos mais expressivos do valor econômico gerado, especialmente nas décadas mais recentes. Esse comportamento indica que, mesmo em contextos de relativa estabilidade produtiva, a atividade mantém sua relevância econômica, sugerindo ganhos associados à valorização do produto, maior inserção comercial ou mudanças na estrutura de mercado.

- Nos anos mais recentes da série, observa-se um aumento mais acentuado do valor da produção, atingindo os maiores patamares do período analisado. Tal comportamento reforça a interpretação de que a cajucultura permanece economicamente significativa no contexto agropecuário nacional, ainda que não apresente crescimento proporcional em termos de volume físico.

- De forma geral, a análise do valor da produção demonstra que a avaliação do desempenho da cajucultura não pode ser realizada exclusivamente com base em indicadores quantitativos de produção. O valor econômico emerge como variável central para compreender a dinâmica da atividade, evidenciando que a relevância do setor está associada não apenas à quantidade produzida, mas à capacidade de gerar renda e valor ao longo do tempo.

# Síntese analítica

O valor da produção apresenta tendência de crescimento no longo prazo;
As oscilações interanuais indicam sensibilidade econômica da atividade;
O desempenho econômico não acompanha necessariamente o comportamento da produção física;
A cajucultura mantém relevância econômica contínua, especialmente nas décadas mais recentes.

```python
# Eficiência da cajucultura no Brasil
eficiencia_cajucultura = cajucultura_brasil.merge(
    valor_cajucultura_brasil,
    on="ano",
    how="inner"
)
```

```python
# Indicador de valor por unidade produzida (R$/tonelada)
eficiencia_cajucultura["valor_por_unidade"] = (
    eficiencia_cajucultura["valor_total_producao"] /
    eficiencia_cajucultura["producao_total"]
)
eficiencia_cajucultura.head(10)
```

```python
# Gráfico da eficiência da cajucultura no Brasil
import plotly.express as px

fig = px.line(
    eficiencia_cajucultura,
    x="ano",
    y="valor_por_unidade",
    title="Eficiência Econômica Implícita da Cajucultura no Brasil",
    labels={
        "ano": "Ano",
        "valor_por_unidade": "Valor por unidade produzida (R$/unidade)"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()
```

# Eficiência econômica implícita da cajucultura no Brasil

-A evolução da eficiência econômica implícita da cajucultura no Brasil evidencia uma trajetória de valorização progressiva da atividade ao longo do tempo, expressa pelo aumento do valor gerado por unidade física produzida. Observa-se que, a partir da década de 1990, o indicador deixa patamares próximos de zero e passa a apresentar crescimento gradual, acompanhando a consolidação da castanha de caju como principal referência econômica da cadeia produtiva.

-Ao longo dos anos 2000, a eficiência implícita apresenta oscilações moderadas, porém mantém uma tendência ascendente, indicando que, mesmo em contextos de estabilidade ou crescimento limitado da produção física, a atividade passou a gerar maior retorno econômico por unidade produzida. Esse comportamento sugere um processo de valorização do produto, associado a fatores de mercado, preços e maior relevância econômica da cajucultura no cenário agropecuário nacional.

-A partir da década de 2010, observa-se uma aceleração mais expressiva do indicador, com elevação consistente do valor por unidade produzida e alcance dos maiores patamares da série histórica nos anos mais recentes. Embora ocorram oscilações interanuais, o nível médio da eficiência econômica implícita permanece significativamente superior ao observado em períodos anteriores, evidenciando uma mudança estrutural no desempenho econômico da atividade.

- De forma geral, o gráfico demonstra que a cajucultura passou por um processo de descolamento entre produção física e geração de valor, no qual ganhos econômicos ocorreram independentemente de aumentos proporcionais na quantidade produzida. Esse resultado reforça a importância de análises que considerem simultaneamente indicadores físicos e econômicos, uma vez que a eficiência implícita emerge como elemento central para a compreensão da dinâmica e da sustentabilidade econômica da cadeia produtiva ao longo do tempo.
