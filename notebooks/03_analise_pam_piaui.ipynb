# Como a cajucultura evoluiu no Piauí?

# O comportamento estadual acompanha o padrão nacional?

# O Piauí gera valor ou apenas volume?

# Há sinais de eficiência econômica implícita?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

plt.rcParams['figure.figsize'] = (10, 6)
```

```python
piaui = pd.read_csv(
    "../data/raw/Pesquisa Agrícola Municipal (PAM)/5. caju_castanha_piaui_desempenho_geral.csv"
)
piaui.head()
```

```python
piaui_limpo = piaui.dropna(subset=["producao_total", "valor_total_producao"])

piaui_limpo["produto"].value_counts()
```

```python
piaui_serie = (
    piaui_limpo
    .groupby("ano", as_index=False)
    .agg({
        "area_colhida_ha": "sum",
        "producao_total": "sum",
        "valor_total_producao": "sum"
    })
)
piaui_serie.head(10)

```

```python
fig = px.line(
    piaui_serie,
    x="ano",
    y="producao_total",
    title="Evolução da Produção da Cajucultura no Piauí",
    labels={
        "ano": "Ano",
        "producao_total": "Produção total da cajucultura"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()
```

# Evolução da produção da cajucultura no Piauí

- A evolução da produção da cajucultura no estado do Piauí apresenta um comportamento marcado por duas fases claramente distintas ao longo do período analisado. Na primeira fase, concentrada até o final da década de 1980, observa-se um crescimento expressivo da produção, culminando em um pico elevado em meados da década, quando os volumes produzidos atingem os maiores patamares da série histórica.

- Esse período inicial é caracterizado por forte variabilidade interanual e por valores absolutos significativamente superiores aos observados nas décadas posteriores. O comportamento sugere um modelo de registro da produção distinto daquele adotado nos anos mais recentes, no qual os volumes contabilizados refletem uma concepção mais ampla da atividade produtiva.

- A partir do final da década de 1980, verifica-se uma redução abrupta dos valores de produção, seguida por uma trajetória de níveis substancialmente menores e maior estabilidade relativa. Essa mudança representa uma inflexão estrutural na série temporal, indicando uma alteração na forma de mensuração da produção, e não necessariamente um desaparecimento da atividade produtiva no estado.

- Nas décadas seguintes, a produção da cajucultura no Piauí mantém-se contínua, porém em patamares reduzidos, com oscilações interanuais moderadas. Esse comportamento evidencia a permanência da atividade no estado, ainda que com menor expressão em termos de volume físico, quando comparada ao período inicial da série.

- De forma geral, o gráfico indica que a análise da produção da cajucultura no Piauí exige cautela na comparação direta entre diferentes períodos históricos. A presença de uma ruptura clara na escala dos valores reforça a necessidade de considerar a série sob a perspectiva da continuidade da atividade, e não apenas a partir dos níveis absolutos de produção, especialmente em análises comparativas e avaliações de desempenho de longo prazo.

# Síntese dos resultados observados

A produção apresenta pico expressivo até o final da década de 1980;
Há uma quebra abrupta nos valores a partir desse período;
O período posterior é marcado por continuidade com menor escala produtiva;
A cajucultura permanece presente no estado, ainda que com menor volume físico.

```python
fig = px.line(
    piaui_serie,
    x="ano",
    y="valor_total_producao",
    title="Evolução do Valor da Produção da Cajucultura no Piauí",
    labels={
        "ano": "Ano",
        "valor_total_producao": "Valor total da produção (R$)"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()
```

# Valor da produção da cajucultura no Piauí

- A evolução do valor da produção da cajucultura no estado do Piauí evidencia uma trajetória marcada por continuidade econômica da atividade, ainda que associada a oscilações interanuais relevantes. A partir da década de 1990, o valor da produção passa a ser registrado de forma mais consistente, acompanhando a consolidação da castanha de caju como principal referência econômica da cadeia produtiva estadual.

- Ao longo dos anos 2000, observa-se um comportamento caracterizado por variações frequentes no valor gerado, com alternância entre períodos de crescimento e retração. Essas oscilações indicam sensibilidade da atividade a fatores conjunturais, como condições de mercado, preços e possíveis restrições produtivas, sem que isso implique descontinuidade da geração de valor.

- Nas décadas mais recentes, especialmente a partir de meados dos anos 2010, o gráfico revela uma tendência de elevação do valor da produção, atingindo os maiores patamares da série histórica nos anos finais do período analisado. Esse comportamento sugere um processo de valorização econômica da cajucultura no estado, mesmo em um contexto de produção física relativamente limitada.

- De forma geral, os dados indicam que o desempenho econômico da cajucultura no Piauí não está diretamente condicionado a aumentos expressivos de volume produtivo. A manutenção e ampliação do valor gerado ao longo do tempo reforçam a interpretação de que a atividade possui relevância econômica contínua, associada à capacidade de geração de renda, e não apenas à quantidade produzida.

# Síntese dos resultados observados

O valor da produção apresenta continuidade temporal desde os anos 1990;
Há oscilações interanuais significativas, refletindo sensibilidade econômica;
Os maiores valores concentram-se nos anos mais recentes da série;
A relevância econômica da cajucultura no Piauí persiste mesmo com baixa produção física.

```python
eficiencia_piaui = piaui_serie.copy()

eficiencia_piaui["valor_por_unidade"] = (
    eficiencia_piaui["valor_total_producao"] /
    eficiencia_piaui["producao_total"]
)
```

```python
fig = px.line(
    eficiencia_piaui,
    x="ano",
    y="valor_por_unidade",
    title="Eficiência Econômica Implícita da Cajucultura no Piauí",
    labels={
        "ano": "Ano",
        "valor_por_unidade": "Valor por unidade produzida (R$/unidade)"
    }
)

fig.update_layout(
    template="plotly_dark",
    title_x=0.5
)

fig.show()
```

- A eficiência econômica implícita da cajucultura no estado do Piauí evidencia uma trajetória clara de aumento do valor gerado por unidade produzida ao longo do período analisado. Observa-se que, até o início da década de 1990, o indicador permanece próximo de zero, refletindo a ausência ou baixa expressividade econômica da produção mensurada nesse período.

- A partir da década de 1990, a eficiência implícita passa a apresentar crescimento gradual, indicando que a cajucultura no estado começa a gerar retornos econômicos mais consistentes por unidade de produção. Esse movimento ocorre mesmo em um contexto de produção física relativamente reduzida, evidenciando um descolamento entre volume produzido e valor econômico gerado.

- Nos anos 2000, o indicador mantém uma tendência de elevação com oscilações moderadas, sugerindo que, embora haja variações interanuais, o nível médio de eficiência permanece superior ao observado nas décadas anteriores. Esse comportamento indica um processo de consolidação econômica da atividade, no qual a geração de valor se torna menos dependente do aumento da produção física.

- A partir da década de 2010, observa-se uma aceleração mais expressiva da eficiência econômica implícita, com o indicador atingindo os maiores patamares da série histórica nos anos mais recentes. Apesar de oscilações pontuais, o valor por unidade produzida mantém-se elevado, reforçando a interpretação de que a cajucultura no Piauí passou por um processo de valorização econômica ao longo do tempo.

- De forma geral, os dados indicam que o desempenho da cajucultura no Piauí está cada vez mais associado à capacidade de geração de valor econômico, e não apenas à expansão da produção física. A eficiência econômica implícita emerge, portanto, como elemento central para a compreensão da sustentabilidade e da relevância da atividade no contexto estadual.
